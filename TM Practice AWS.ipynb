```python
#removing the lower function from line to enable POS tagging to work better. is there a better fix?
def tokenize_fast(line):
    import re
    # tokenize using reg ex (fast)
    tokens = re.findall("[A-Z]{2,}(?![a-z])|[A-Z][a-z]+(?=[A-Z])|[\'\w\-]+",line)
    # remove punctuation on either end
    from string import punctuation
    tokens = [tok.strip(punctuation) for tok in tokens]
    # make sure each thing in list isn't empty
    tokens = [tok for tok in tokens if tok]
    return tokens
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
# a function to tokenize and remove stopwords / small words
def tokenize_pos_lower (text, pos='NN', stop_words=stopwords.words('english')):
    tokens = tokenize_fast(text)
    tagged_tokens = nltk.pos_tag(tokens)
   # tokens2=[]
    #    for w,p in tagged_tokens:
     #   if p == pos:
      #  tokens2.append(w)
       # tokens2=[]
    tokens2 = [
        w
        for w,p in tagged_tokens
        if p==pos and len(w)>3 and w.lower not in stop_words
    ]   
    return tokens2
```

```python
from tqdm import tqdm

texts2 = [tokenize_pos_lower(t.txt) for t in tqdm(C.texts())]
```

```python
from gensim import corpora

dictionary = corpora.Dictionary(texts)
dictionary.filter_extremes(no_below=5, no_above=0.7, keep_n=4000) #paramaters to tweak
corpus = [dictionary.doc2bow(text) for text in texts2]
```

```python
from gensim import models

n_topics = 18

lda_model = models.LdaModel(corpus=corpus, num_topics=n_topics)
```

## Increasing no_above results in tokens that are more common across the whole corpus
## Decreasing no_above_ results in tokens that are less common

```python
import pyLDAvis
import pyLDAvis.gensim_models
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
vis
```
