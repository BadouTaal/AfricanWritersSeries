## Let's try to compare corpora - AWS vs Chicago

```python
dtm1 = C.dtm()
```

```python
dtm2 = C2.dtm()
```

```python
dtm1['corpus']='AWS'
dtm2['corpus']='CHI'
```

```python
dtm = dtm1.append(dtm2)
```

```python
balanced = dtm.groupby('corpus').sample(n=100)
```

```python
from lltk.model.classifier import Classifier
clf = Classifier(balanced1)
```

```python
clf.classify('corpus')
```

```python
clf.report()
```

```python
clf.dfr
```

```python
clf.dfc
```

```python
topfeats_df = clf.dfc.head(25).set_index('feat')
topfeats_df
```
