## Some preprocessing functions

```python
import gensim
from tqdm import tqdm
```

```python
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            nltk.bigrams(token)
            result.append((token))
    return result
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def tokenize_pos(text, pos='NN', stop_words=stopwords.words('english')):
    tokens = tokenize_fast(text)
    tagged_tokens = nltk.pos_tag(tokens)
   # tokens2=[]
    #    for w,p in tagged_tokens:
     #   if p == pos:
      #  tokens2.append(w)
       # tokens2=[]
    tokens2 = [
        w
        for w,p in tagged_tokens
        if p==pos and len(w)>3 and w not in stop_words
    ]   
    return tokens2
```
